```python
import dash
from dash import dcc, html, Input, Output

# Initialize the Dash app (no longer need JupyterDash)
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Genomic Testing Workflow Dashboard"),
    dcc.Dropdown(
        id='color-dropdown',
        options=[
            {'label': 'Red', 'value': 'Red'},
            {'label': 'Blue', 'value': 'Blue'},
            {'label': 'Green', 'value': 'Green'}
        ],
        value='Red'
    ),
    html.Div(id='output-message')
])

@app.callback(
    Output('output-message', 'children'),
    Input('color-dropdown', 'value')
)
def update_output(color):
    return 'You selected ' + color

if __name__ == '__main__':
    # Use app.run with jupyter_mode instead of JupyterDash's run_server
    app.run(jupyter_mode='inline')
```
